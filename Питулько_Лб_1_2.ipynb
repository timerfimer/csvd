```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/shopping_trends.csv", encoding='cp1251')

print(df.head())
```

```python
print("Розмір таблиці:", df.shape)

print("\nТипи стовпців:")
print(df.dtypes)

print("\nКількість пропусків:")
print(df.isnull().sum())
```

```python

sns.histplot(data=df, x='Age', bins=20, kde=True)

plt.title('Гістограма віку клієнтів')
plt.xlabel('Вік')
plt.ylabel('Кількість клієнтів')

plt.show()
```

```python
top_items = df['Item Purchased'].value_counts().reset_index()

top_items.columns = ['Товар', 'Кількість покупок']

plt.figure(figsize=(12, 6))

sns.barplot(data=top_items.head(10), x='Кількість покупок', y='Товар', palette='bright')

plt.title("Найпопулярніші товари")
plt.xlabel('Кількість покупок')
plt.ylabel('Товар')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Age', y='Purchase Amount (USD)', alpha=0.6, color='green')


plt.title('Залежність між віком та сумою покупок')

plt.xlabel('Вік')

plt.ylabel('Сума покупок (USD)')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Review Rating', bins=9, kde=True, color='red')

plt.title('Розподіл оцінок товарів')
plt.xlabel('Оцінка товару')
plt.ylabel('Кількість оцінок')

plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Gender', palette='muted')

plt.title('Розподіл покупок за статтею')

plt.xlabel('Стать')

plt.ylabel('Кількість покупок')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Review Rating', y='Purchase Amount (USD)', alpha=0.6, color='r')
plt.title('Залежність між рейтингом товару та сумою покупок')
plt.xlabel('Рейтинг товару')
plt.ylabel('Сума покупок USD')

plt.show()
```

```python
season_counts = df['Season'].value_counts().reset_index()
season_counts.columns = ['Сезон', 'Кількість покупок']
season_order = ['Winter', 'Spring', 'Summer', 'Fall']
plt.figure(figsize=(12,6))
sns.barplot(data=season_counts, x='Сезон', y='Кількість покупок', order=season_order, palette='colorblind')
plt.title('Сезонність покупок')
plt.xlabel('Сезон')
plt.ylabel('Кількість покупок')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
size_order = ['S', 'M', 'L', 'XL']
sns.countplot(data=df, x='Size', order=size_order)
plt.title('Розподіл розміру одягу')
plt.xlabel('Розмір')
plt.ylabel('Кількість покупок')

plt.show()
```

```python
payment_counts = df['Payment Method'].value_counts().reset_index()
payment_counts.columns = ['Метод оплати', 'Кількість покупок']

plt.figure(figsize=(12,6))
sns.barplot(data=payment_counts, x='Метод оплати', y='Кількість покупок', palette='tab20c')
plt.title('Кількість покупок за методом оплати')
plt.xlabel('Методи оплати')
plt.ylabel('Кількість покупок')

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Previous Purchases', y='Purchase Amount (USD)', alpha=0.6, color='grey')
plt.title('Залежність між кількість попередніх покупок і сумою поточної покупки')
plt.xlabel('Кількість покупок')
plt.ylabel('Сума поточної покупки USD')

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z)
ax.set_title('Графік поверхні')
ax.set_xlabel('Вісь X')
ax.set_ylabel('Вісь Y')
ax.set_zlabel('Вісь Z')
fig.colorbar(surf)

plt.show()
```

```python
data_matrix = np.random.rand(4, 4)

plt.figure(figsize=(10, 6))
sns.heatmap(data_matrix, annot=True, fmt=".2f")
plt.title('Теплова поверхня HEAT MAP')
plt.xlabel('Ось Х')
plt.ylabel('Ось Y')

plt.show()
```

Висновок:У цій лабораторній роботі було проведено аналіз даних з файлу shopping_trends.csv. Візуалізували залежність між віком та сумою покупок за допомогою scatter plot.Побудували графік поверхні для візуалізації залежності між трьома числовими змінними з використанням бібліотеки Seaborn, а також графік теплової поверхні для візуалізації залежності між двома числовими змінними на основі матриці даних.
